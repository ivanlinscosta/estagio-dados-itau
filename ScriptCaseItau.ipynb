```python
%pip install Jinja2
#importing the libraries used
import pandas as pd 
import matplotlib.pyplot as plt

#funções utilizadas
principal_table = pd.read_csv("Ecommerce_DBS.csv") #importando a tabela principal

#calcula o NPS de uma categoria de produto
def calculonps (produto):
    principal_table = pd.read_csv("Ecommerce_DBS.csv") #lendo a tabela principal
    principal_table = principal_table[['Product Category','NPS']].loc[principal_table["Product Category"] == produto].dropna() #criando uma tabela filtrada apenas com cada venda e seu nps
    NPS_Count = principal_table['NPS'].value_counts().to_frame().reset_index().rename(columns={'index': 'NPS', 'NPS': 'Quantidade'}).sort_values(by='NPS') #counting the number of notes in each sale
    print('Calculo do NPS da categoria:', produto)
    print('\n\n')
    display(NPS_Count)
    print('\n\n')
    #separando as faixas de valores para o calculo do NPS 
    faixas=labels=['0 a 6' , '7 a 8', '9 a 10']
    NPS_Count['Faixas Numéricas'] = pd.cut(x=NPS_Count['NPS'] , bins=[0,6,8,10], labels = faixas)
    NPS_Count = NPS_Count.groupby(['Faixas Numéricas']).sum()
    display(NPS_Count)
    print('\n\n')
    #calculo do NPS
    total = NPS_Count['Quantidade'].sum() #total
    q1 = NPS_Count['Quantidade'][0] #faixa de 0 a 6
    q2 = NPS_Count['Quantidade'][2] #faixa de 9 a 10 
    NPS = (q2/total - q1/total) * 100
    print("O NPS dessa categoria de produtos é:", NPS)
    return 

#Cria tabelas para analisar a quantidade de vendas e o faturamento total de cada categoria por genero
def source_and_gender(product): 
    #filtrando acolunas e linhas 
    electronic_instagram = principal_table[['Product Category','Quantity','Gender','Source', 'Total Purchase Amount']].loc[principal_table["Product Category"] == product].loc[principal_table["Source"]== "Instagram Campign"].groupby(['Product Category', 'Gender', 'Source']).sum().dropna()
    electronic_facebook = principal_table[['Product Category','Quantity','Gender','Source', 'Total Purchase Amount']].loc[principal_table["Product Category"] == product].loc[principal_table["Source"]== "FaceBook campaign"].groupby(['Product Category', 'Gender', 'Source']).sum().dropna()
    display(electronic_instagram)
    print('\n\n')
    display(electronic_facebook)
    print('\n\n')
    return

#Gera um gráfico da quantidade de vendas ou do faturamento financeiro, filtrando as vendas por Categoria de Produto, Canal de Venda e gênero.
def grafico(Product, Source, Gender, acumulo, titulo):
    table_filtrada = principal_table.loc[principal_table["Product Category"] == Product].loc[principal_table['Source']== Source].loc[principal_table["Gender"]== Gender] #filtering the rows
    table_filtrada1 = table_filtrada[[ 'Customer Age ' , acumulo]].groupby(['Customer Age ']).sum().dropna()#filtering the columns and grouping ages
    plt.plot(table_filtrada1, marker='o', linestyle='-')
    plt.xlabel('Idade')
    plt.ylabel('Quantidade')
    plt.title(titulo)
    return plt.show()

#gera as medidas de dispersão de centralidade para uma categoria de produtos 
def medidas(categoria):
    table_prod = principal_table[["Product Category", "Product Price"]].loc[principal_table["Product Category"]== categoria]
    print('As medidas de centralidade e dispersão do produto são:', table_prod.describe())
    print('Mediana:', table_prod['Product Price'].median())
    

def main():
    #1- Produtos mais vendidos considerando os últimos 3 anos
    print("1- Produtos mais vendidos considerando os últimos 3 anos\n\n\n")
    principal_table["Purchase Date"] = pd.to_datetime(principal_table["Purchase Date"] , dayfirst=True) #mudando as datas da tabela para o formato "data time"
    data_table1 = principal_table[["Purchase Date", "Quantity", "Product Category"]] #tabela com colunas filtradas
    init_date = '2020-01-01' #definindo o inicio do periodo analise
    end_date = '2022-12-31'  #definindo o final do periodo de analise
    data_table2 = data_table1[(data_table1['Purchase Date'] >= init_date) & (data_table1['Purchase Date'] <= end_date)].dropna().groupby(["Product Category"]).sum("Quantity").sort_values(by='Quantity',ascending=False) #filtrando e agrupando as compras para o periodo completo de 3 anos, por categoria
    
    display(data_table2.style.highlight_max()) #exibindo a tabela
    print('\n\n')
    
    #criação de um grafico de barras para visualização dos dados
    cor = ('#FFA500')
    data_table2.plot(kind ='bar' , color= cor, title = 'QUANTIDADE DE VENDA NOS ÚLTIMOS 3 ANOS' ) 
    
    
    
    
    
    # 2- Produto mais caro e o mais barato?
    print('\n\n\n\nProduto mais caro e o mais barato\n\n')
    value_table = principal_table[["Product Category", "Product Price"]].dropna().drop_duplicates().sort_values(by='Product Price',ascending=False) #criando uma tabela com as categorias de produtos e seus preços, ordenando de forma decrescente
    produto_mais_caro = principal_table["Product Price"].max() #obtendo o valor do produto mais caro
    produto_mais_barato = principal_table["Product Price"].min() #obtendo o valor do produto mais barao
    print("\n\nO produto mais caro custa R$:", produto_mais_caro , "\nO produto mais barato custa R$:", produto_mais_barato)
    print('\n\n')
    value_table
    
    
    
    
    #3- Categoria de produto mais vendida e menos vendida e Categoria mais e menos cara?
    print('\n\n\n\nCategoria de produto mais vendida e menos vendida e Categoria mais e menos cara\n\n')
    
    #Tabela que mostra a quantidade de vendas por categoria
    table_product_category = principal_table[["Product Category" , "Quantity"]].groupby(["Product Category"]).sum().sort_values(by='Quantity',ascending=False)
    display(table_product_category.style.highlight_max())
    print('\n\n')
    #grafico que mostra a quantidade de venda por categoria
    table_product_category.plot(kind ='bar' , color=cor, title = 'QUANTIDADE DE VENDAS POR CATEGORIA')
    
    
    #Medidas de dispersão e centralidade por categoria
    medidas('Electronics')
    print('\n\n')
    medidas('Books')
    print('\n\n')
    medidas('Home')
    print('\n\n')
    medidas('Clothing')
    print('\n\n\n\n')
    
    #tabelas que mostram os valores unicos de produtos 
    table_elec_ = value_table.loc[principal_table["Product Category"] == 'Electronics'].dropna()
    table_books_ = value_table.loc[principal_table["Product Category"] == 'Books'].dropna()
    table_home_ = value_table.loc[principal_table["Product Category"] == 'Home'].dropna()
    table_clothing_ = value_table.loc[principal_table["Product Category"] == 'Clothing'].dropna()
    display(table_elec_)
    print('\n\n')
    display(table_books_)
    print('\n\n')
    display(table_home_)
    print('\n\n')
    display(table_clothing_)
    
    
    
    #4 NPS
    print('\n\n\n\nCalculo do NPS\n\n')
    calculonps('Electronics');
    print('\n\n\n')
    calculonps('Home');
    print('\n\n\n')
    calculonps('Clothing');
    print('\n\n\n')
    calculonps('Books');
    
    
    print('\n\n\n\nAnalise dos melhores públicos\n\n\n')
    ## ANALISE GENERO- ELECTRONICS 
    print('\n\n\n\nCATEGORIA DE PRODUTOS ELETRONICOS\n\n')
    #tabela para analisar a quantidade de vendash e o faturamento total de cada categoria por genero
    source_and_gender('Electronics');
    print('\n\n\n')
    #Criação de um grafico de pizza 
    labels = ['Female Instagram Costumers', 'Male Instagram Costumers', 'Female Facebook Costumers', 'Male Facebook Costumers'] #títulos
    quantidades = [21564, 21222, 19962, 20351] #valores atrelados aos titulos
    colors = ['#DB7093' , '#00008B', '#FFB6C1', '#ADD8E6'] #cores
    plt.figure(figsize=(3, 3)) 
    plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140) 
    plt.title('Distribuição de gênero - Produtos eletrônicos') 
    plt.axis('equal') 
    plt.show() 
    
    
    #ANALISE GENERO- ELECTRONICS/FEMALE
    
    grafico('Electronics', 'Instagram Campign', 'Female', 'Quantity',  'Grafico Quantidade- Instagram, Público feminino');
    grafico('Electronics', 'Instagram Campign', 'Male', 'Quantity', 'Grafico Quantidade- Instagram, Público masculino');
    grafico('Electronics', 'FaceBook campaign', 'Female' , 'Quantity' , 'Grafico Quantidade- Facebook, Público feminino');
    grafico('Electronics', 'FaceBook campaign', 'Male', 'Quantity', 'Grafico Quantidade- Facebook, Público masculino');
    grafico('Electronics', 'Instagram Campign', 'Female', 'Total Purchase Amount',  'Grafico Faturamento- Instagram, Público feminino');
    
    
    
    
    
    #ANALISE GENERO- BOOKS
    print('\n\n\n\nCATEGORIA DE PRODUTOS DE LIVROS\n\n')
    #tabelas
    source_and_gender('Books');
    print('\n\n\n')
    
    #grafico pizza
    labels = ['Female Instagram Costumers', 'Male Instagram Costumers', 'Female Facebook Costumers', 'Male Facebook Costumers']
    quantidades = [31316, 31153, 29413, 30145] 
    colors = ['#DB7093' , '#00008B', '#FFB6C1', '#ADD8E6'] 
    plt.figure(figsize=(3, 3)) 
    plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140) 
    plt.title('Distribuição do gênero - Produtos de Livros') 
    plt.axis('equal') 
    plt.show() 
    
    #graficos linha
    grafico('Books', 'Instagram Campign', 'Female', 'Quantity', 'Grafico Quantidade- Instagram, Público feminino');
    grafico('Books', 'Instagram Campign', 'Male', 'Quantity', 'Grafico Quantidade- Instagram, Público masculino');
    grafico('Books', 'FaceBook campaign', 'Female', 'Quantity', 'Grafico Quantidade- Facebook, Público feminino');
    grafico('Books', 'FaceBook campaign', 'Male', 'Quantity', 'Grafico Quantidade- Facebook, Público masculino');
    grafico('Books', 'Instagram Campign', 'Female', 'Total Purchase Amount',  'Grafico Faturamento- Instagram, Público feminino');
    
    
    
    
    #ANALISE GENERO- HOME
    print('\n\n\n\nCATEGORIA DE PRODUTOS DE CASA\n\n')
    #Tabelas
    source_and_gender('Home');
    print('\n\n\n')
    #grafico pizza
    labels = ['Female Instagram Costumers', 'Male Instagram Costumers', 'Female Facebook Costumers', 'Male Facebook Costumers'] #títulos
    quantidades = [21288, 20964, 19835, 19628] #values linked to titles
    colors = ['#DB7093' , '#00008B', '#FFB6C1', '#ADD8E6'] #colors
    plt.figure(figsize=(3, 3)) #grafic size
    plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140) #creating the grafic
    plt.title('Distribuição de gênero - Produtos de Casa') #title
    plt.axis('equal') #formatation
    plt.show() #exhibition
    
    #graficos linha
    grafico('Home', 'Instagram Campign', 'Female', 'Quantity', 'Grafico Quantidade- Instagram, Público feminino');
    grafico('Home', 'Instagram Campign', 'Male', 'Quantity', 'Grafico Quantidade- Instagram, Público masculino');
    grafico('Home', 'FaceBook campaign', 'Female', 'Quantity', 'Grafico Quantidade- Facebook, Público feminino');
    grafico('Home', 'FaceBook campaign', 'Male', 'Quantity', 'Grafico Quantidade- Facebook, Público masculino');
    grafico('Home', 'Instagram Campign', 'Female', 'Total Purchase Amount',  'Grafico Faturamento- Instagram, Público feminino');
    
    
    
    #ANALISE GENERO- CLOTHING
    print('\n\n\n\nCATEGORIA DE PRODUTOS DE ROUPA\n\n')
    #tabelas
    source_and_gender('Clothing');
    
    #grafico pizza
    labels = ['Female Instagram Costumers', 'Male Instagram Costumers', 'Female Facebook Costumers', 'Male Facebook Costumers'] #títulos
    quantidades = [31590, 31376, 30624, 29542] #values linked to titles
    colors = ['#DB7093' , '#00008B', '#FFB6C1', '#ADD8E6'] #colors
    plt.figure(figsize=(3, 3)) #grafic size
    plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140) #creating the grafic
    plt.title('Distribuição de gênero - Roupas') #title
    plt.axis('equal') #formatation
    plt.show() #exhibition
    
    #graficos linha
    grafico('Clothing', 'Instagram Campign', 'Female', 'Quantity', 'Grafico Quantidade- Instagram, Público feminino');
    grafico('Clothing', 'Instagram Campign', 'Male', 'Quantity', 'Grafico Quantidade- Instagram, Público masculino');
    grafico('Clothing', 'FaceBook campaign', 'Female', 'Quantity', 'Grafico Quantidade- Facebook, Público feminino');
    grafico('Clothing', 'FaceBook campaign', 'Male', 'Quantity', 'Grafico Quantidade- Facebook, Público masculino');
    grafico('Clothing', 'Instagram Campign', 'Female', 'Total Purchase Amount',  'Grafico Faturamento- Instagram, Público feminino');

if __name__=="__main__":
    main()
```
