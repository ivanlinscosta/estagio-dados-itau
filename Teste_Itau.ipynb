# About Dataset
###The data set contains the following columns:
- Customer ID is a Unique ID assigned to each customer based on the country.
- Purchase Date: The date of purchase for a customer is in a short date format.
- Product Category: The category of product on the website.
- Product Price: The price of the purchased item.
- Quantity: The number of products that customers bought
- Total Purchase Amount: The purchase since the current purchase.
- NPS: The Satisfaction factor of the customer (0 the worst to 10 the best)
- Customer Age: This field is a numerical field range from 18-70.
- Gender: Male/ Female.
- Source: the source where the customer is redirected to the website.
- Country: Customer’s country. (The United States/ Canada)
- State: Customer state.
- Latitude: Latitude of living state
- Longitude: Longitude of living state.

##3 - Qual a categoria de produto mais vendida e menos vendida? Qual a categoria mais e menos cara?

```python
# Determinando a categoria mais vendida e menos vendida
category_sales = df_clone.groupby('ProductCategory')['Quantity'].sum().reset_index()
most_sold_category = category_sales.loc[category_sales['Quantity'].idxmax()]
least_sold_category = category_sales.loc[category_sales['Quantity'].idxmin()]

print("Categoria mais vendida:")
print(most_sold_category)

print("\nCategoria menos vendida:")
print(least_sold_category)

# Determinando a categoria mais cara e menos cara com base na quantidade de itens caros e baratos
# Aqui definimos "caro" como acima da mediana dos preços e "barato" como abaixo da mediana
price_median = df_clone['ProductPrice'].median()

# Criando colunas auxiliares para identificar itens caros e baratos
df_clone['Expensive'] = df_clone['ProductPrice'] > price_median
df_clone['Cheap'] = df_clone['ProductPrice'] <= price_median

# Contando itens caros e baratos por categoria
expensive_count = df_clone.groupby('ProductCategory')['Expensive'].sum().reset_index()
cheap_count = df_clone.groupby('ProductCategory')['Cheap'].sum().reset_index()

# Encontrando a categoria com mais itens caros e baratos
most_expensive_category = expensive_count.loc[expensive_count['Expensive'].idxmax()]
cheapest_category = cheap_count.loc[cheap_count['Cheap'].idxmax()]

print("\nCategoria com mais itens caros:")
print(most_expensive_category)

print("\nCategoria com mais itens baratos:")
print(cheapest_category)
```

```python
# Criando gráfico de barras para categorias mais e menos vendidas
fig1 = px.bar(category_sales, x='ProductCategory', y='Quantity',
              title='Quantidade Vendida por Categoria de Produto',
              labels={'ProductCategory': 'Categoria de Produto', 'Quantity': 'Quantidade Vendida'})

# Criando gráfico de barras para itens caros e baratos
expensive_cheap_df = pd.merge(expensive_count, cheap_count, on='ProductCategory')
expensive_cheap_df = expensive_cheap_df.melt(id_vars='ProductCategory',
                                             value_vars=['Expensive', 'Cheap'],
                                             var_name='PriceCategory', value_name='Count')

fig2 = px.bar(expensive_cheap_df, x='ProductCategory', y='Count', color='PriceCategory',
              title='Contagem de Itens Caros e Baratos por Categoria de Produto',
              labels={'ProductCategory': 'Categoria de Produto', 'Count': 'Contagem', 'PriceCategory': 'Tipo de Preço'},
              barmode='group')

# Exibindo os gráficos
fig1.show()
fig2.show()
```

##Respondendo:

Os itens mais caros estão na categoria Books.
Ja os itens menos caros estão na categoria Clothing.

##4 - Qual o produto com melhor e pior NPS?

```python
# Classificando os respondentes
df_clone['Promoters'] = df_clone['NPS'].apply(lambda x: 1 if x >= 9 else 0)
df_clone['Detractors'] = df_clone['NPS'].apply(lambda x: 1 if x <= 6 else 0)
df_clone['Neutral'] = df_clone['NPS'].apply(lambda x: 1 if 7 <= x <= 8 else 0)

# Agrupando por ProductCategory e calculando o NPS
category_nps = df_clone.groupby('ProductCategory').agg({
    'Promoters': 'sum',
    'Detractors': 'sum',
    'Neutral': 'sum',
    'CustomerID': 'count'  # Número total de respondentes
}).reset_index()

category_nps['NPS'] = (category_nps['Promoters'] - category_nps['Detractors']) / category_nps['CustomerID'] * 100

# Encontrando a categoria com o maior e menor valor de NPS
highest_nps_category = category_nps.loc[category_nps['NPS'].idxmax()]
lowest_nps_category = category_nps.loc[category_nps['NPS'].idxmin()]

#print("Categoria com maior valor de NPS:")
print(highest_nps_category)

#print("\nCategoria com menor valor de NPS:")
print(lowest_nps_category)
```

```python
# Calculando as porcentagens
category_nps['Total'] = category_nps['CustomerID']
category_nps['Promoters_pct'] = (category_nps['Promoters'] / category_nps['Total']) * 100
category_nps['Detractors_pct'] = (category_nps['Detractors'] / category_nps['Total']) * 100
category_nps['Neutral_pct'] = (category_nps['Neutral'] / category_nps['Total']) * 100

# Transformando os dados para o formato longo para o gráfico de barras empilhadas
category_nps_long = category_nps.melt(id_vars='ProductCategory',
                                              value_vars=['Promoters_pct', 'Detractors_pct', 'Neutral_pct'],
                                              var_name='NPS_Type', value_name='Percentage')

# Renomeando os valores de NPS_Type para serem mais legíveis
category_nps_long['NPS_Type'] = category_nps_long['NPS_Type'].map({
    'Promoters_pct': 'Promoters',
    'Detractors_pct': 'Detractors',
    'Neutral_pct': 'Neutral'
})

# Criando o gráfico de barras empilhadas com porcentagens
fig = px.bar(category_nps_long, x='ProductCategory', y='Percentage', color='NPS_Type',
             title='Distribuição das Notas de NPS por Categoria de Produto (em %)',
             labels={'ProductCategory': 'Categoria de Produto', 'Percentage': 'Porcentagem (%)', 'NPS_Type': 'Tipo de NPS'},
             text_auto=True)

# Ajustando o layout do gráfico
fig.update_layout(barmode='stack', xaxis_title='Categoria de Produto', yaxis_title='Porcentagem (%)')

# Exibindo o gráfico
fig.show()
```

```python
fig = px.bar(category_nps, x='NPS', y='ProductCategory', orientation='h',
             title='Net Promoter Score (NPS) por Categoria de Produto',
             labels={'NPS': 'NPS (%)', 'ProductCategory': 'Categoria de Produto'})

# Ajustando o layout do gráfico
fig.update_layout(xaxis_title='NPS (%)', yaxis_title='Categoria de Produto')

# Exibindo o gráfico
fig.show()
```

##Resposta


O NPS (NET PROMOTER SCORE) é uma métrica para avaliar o nível de satisfação do cliente e, através de questionários enviados ao consumidor, obtêm-se notas que são categorizadas entre promotores (nota igual ou maior que 9), neutros (entre 7 e 8) e detradores (entre 6 ou abaixo).
Essas categorias indicam pontos de melhorias e pontos de atenção.
Seguindo a fórmula de cálculo do NPS (Promotores-Detradores/Total de Pessoas que Responderam), a categoria de produto com o melhor NPS é a de livros ('Books'). Já a categoria de produto com o pior NPS é de eletrônicos ('Eletronics').

```python
# Função para criar gráficos de barras agrupadas para cada faixa etária
def create_grouped_bar_charts(df, age_group):
    df_filtered = df[df['AgeGroup'] == age_group]

    # Definir as cores para cada fonte
    source_colors = {
'Instagram Campaign': '#0000FF',
        'SEM': '#008000',
        'FaceBook Campaign': '#FFA500',
        'Organic Search': '#800080'
    }

    fig = make_subplots(rows=1, cols=1,
                        subplot_titles=[f'Distribuição de Produtos por Gênero e Canal - Faixa Etária: {age_group}'],
                        specs=[[{"type": "bar"}]])

    for source in df_filtered['Source'].unique():
        df_source = df_filtered[df_filtered['Source'] == source]
        male_color = source_colors[source]
        female_color = f'rgba{tuple(int(male_color.lstrip("#")[i:i+2], 16) for i in (0, 2, 4)) + (0.5,)}'

        fig.add_trace(go.Bar(
            x=df_source['ProductCategory'],
            y=df_source[df_source['Gender'] == 'Male']['Count'],
            name=f'Male - {source}',
            marker_color=male_color,
            marker_pattern_shape="",
            opacity=0.8
        ))
        fig.add_trace(go.Bar(
            x=df_source['ProductCategory'],
            y=df_source[df_source['Gender'] == 'Female']['Count'],
            name=f'Female - {source}',
            marker_color=female_color,
            marker_pattern_shape="\\",
            opacity=0.8
        ))

    fig.update_layout(barmode='group', height=600, width=1000, title_text=f'Faixa Etária: {age_group}')
    fig.show()

# Aplicando a função para cada faixa etária
for age_group in age_order:
    create_grouped_bar_charts(df_count_combined, age_group)
```

```python
# Filtrando para gênero masculino
df_male = df_clone[df_clone['Gender'] == 'Male']

# Criando um DataFrame de contagem para gênero masculino
df_count_male = df_male.groupby(['ProductCategory', 'Source', 'AgeGroup']).size().reset_index(name='Count')

# Criando subplots
fig_male2 = make_subplots(rows=2, cols=2, specs=[[{'type': 'domain'}, {'type': 'domain'}],
                                                 [{'type': 'domain'}, {'type': 'domain'}]],
                          subplot_titles=df_count_male['Source'].unique())

# Função para criar gráficos de pizza para cada ProductCategory e Source
def create_pie_charts(df, product_category):
    sources = df['Source'].unique()

    for i, source in enumerate(sources):
        df_filtered = df[(df['ProductCategory'] == product_category) & (df['Source'] == source)]
        age_counts = df_filtered.groupby(['AgeGroup']).sum().reset_index()

        fig_male2.add_trace(
            go.Pie(labels=age_counts['AgeGroup'], values=age_counts['Count'], name=source),
            row=(i // 2) + 1, col=(i % 2) + 1
        )

    fig_male2.update_traces(hole=.4, hoverinfo="label+percent+name")
    fig_male2.update_layout(
        title_text=f'Distribuição de Produtos por Faixa Etária e Canal - {product_category} (Gênero Masculino)',
        showlegend=False,
        height=800, width=800
    )
    fig_male2.show()

# Aplicando a função para cada categoria de produto
product_categories = df_male['ProductCategory'].unique()
for product_category in product_categories:
    create_pie_charts(df_count_male, product_category)

fig_male2.write_html('distribution_by_age_gender_source_male.html')
```
